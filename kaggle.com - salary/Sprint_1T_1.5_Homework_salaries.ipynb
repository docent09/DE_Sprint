```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('ds_salaries.csv', index_col='Unnamed: 0')
```

**Сгруппируйте вакансии по направлениями (DS, DE, Software Engenering, etc.)**

```python
df.groupby('job_title').salary.count()
```

Видим много разных формулировок вакансий, поэтому создадим дополнительный столбец в DataFrame, в котором будем классифицировать вакансии по ключевым словам.

```python
def job_group_calc(job):
  """Функция, которая определяет принадлежность вакансии к какой-либо группе"""
  if 'Data Scientist' in job or 'Data Science' in job or 'Data Analyst' in job or 'Data Analytics' in job or 'Head of Data' in job:
    return 'DS'
  elif 'Data Engineer' in job or 'Analytics Engineer' in job or 'Data Architect' in job or 'Data Specialist' in job:
    return 'DE'
  elif 'Machine Learning' in job or 'ML ' in job:
    return 'ML'
  elif 'Research Scientist' in job:
    return 'RS'
  elif 'Computer Vision' in job:
    return 'CV'
  elif 'AI Scientist' in job:
    return 'AI'
  else:
    return 'Oth'
```

```python
df['job_group'] = pd.Series([job_group_calc(job) for job in df.job_title.to_list()]) # создаем новую колонку, где будем сохранять тип вакансии
df
```

Делаем группировку по направлениям

```python
df.job_group.value_counts()
```

Вывод: наибольшее количество вакансий по направлению DS - 328 штук.

**Какая средняя и медианная зарплата по группам вакансий?**

```python
df_group = df.groupby('job_group')['salary_in_usd'].agg(['mean', 'median'])
df_group
```

Вывод: найдены средняя и медианная зарплата по группам вакансий

**Какая средняя и медианная зарплата по каждому региону?**

```python
df.company_location.unique()
```

Видим много разных стран. Сгруппируем страны по регионам и создадим новую колонку в DataFrame.

```python
def region_calc(country):
  """Функция, определяющая регион страны по ее обозначению"""
  if country in ['DE', 'GB', 'HU', 'FR', 'GR', 'NL', 'AT', 'ES', 'PT', 'DK', 'IT', 'HR', 'LU', 'PL', 'RO', 'BE', 'UA', 'RU', 'MT', 'MD', 'SI', 'CH', 'CZ', 'EE', 'IE']:
    return 'Europe'
  elif country in ['HN', 'US', 'MX', 'CA']:
    return 'N.America'
  elif country in ['BR', 'CL', 'CO']:
    return 'S.America'
  elif country in ['JP', 'IN', 'PK', 'CN', 'AE', 'SG', 'IQ', 'IL', 'IR', 'VN', 'TR', 'MY']:
    return 'Asia'
  elif country in ['NZ', 'AS', 'AU']:
    return 'Australia'
  elif country in ['NG', 'KE', 'DZ']:
    return 'Africa'
  else:
    return 'Other'
```

```python
df['region'] = pd.Series([region_calc(country) for country in df.company_location.to_list()]) # создаем новую колонку, где будем сохранять регион
df
```

Находим среднюю и медианную зарплату по регионам

```python
df_region = df.groupby('region')['salary_in_usd'].agg(['mean', 'median'])
df_region
```

Вывод: найдены средняя и медианная зарплата по региону расположения компании

**Какая самая высокооплачиваемая из групп вакансий, исходя из их средних зарплат?**

```python
max_salary_group = df_group[df_group['mean']==df_group['mean'].max()]
max_salary_group
```

Вывод: самая высокооплачиваемая группа вакансий - DE.

**Какое процентное соотношение каждого региона по вакансиям от всех вакансий?**

```python
df_region_count = df.groupby('region').region.count()
df_region_percent = round(100 * df_region_count / df_region_count.sum(), 2)
df_region_percent
```

Вывод: наибольший процент от общего количества вакансий в Северной Америке.

**Какая корреляция уровня опыта от зарплаты?**

Создаем вспомогательную таблицу с двумя столбцами

```python
df_exp = pd.DataFrame(df, columns=['experience_level', 'salary_in_usd'])
df_exp.experience_level.unique()
# EN Entry-level / Junior MI Mid-level / Intermediate SE Senior-level / Expert EX Executive-level 
```

```python
def rank_exp(exp_level):
  """Ранжирование уровней опыта"""
  if exp_level == 'EN':
    return 1
  elif exp_level == 'MI':
    return 2
  elif exp_level == 'SE':
    return 3
  elif exp_level == 'EX':
    return 4
  else:
    return None
```

```python
df_exp['exp_lvl'] = pd.Series([rank_exp(exp_level) for exp_level in df_exp.experience_level.to_list()]) # создаем новую колонку, где будем сохранять ранжированный опыт
df_exp
```

```python
df_exp.corr(method='spearman')
```

Коэффициент ранговой корреляции Спирмена между уровнем опыта и зарплатой = 0,56
